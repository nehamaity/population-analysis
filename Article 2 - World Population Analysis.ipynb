# World Population Data Analysis

In this tutorial we will using Python to conduct basic data analysis on a World Population data set.

## Preliminary Steps

Import libraries needed for data analysis

```python
import matplotlib.pyplot as plt
import pandas as pd
```

Read in data from the csv file

```python
popData = pd.read_csv('../Downloads/population_csv.csv')
```

## Explore the Countries & Regions in the Data

```python
popData.drop_duplicates(subset = 'Country Name')
```

## Plotting Multiple Lines on the Same Plot

Now we'll try plotting multiple countries on the same graph to see how their populations grow relative to each other.

```python
pop = popData.loc[(popData['Country Name'] == 'Zimbabwe') | 
(popData['Country Name'] == 'Vietnam') | 
(popData['Country Name'] == 'Spain')]
```

```python
import seaborn as sns
sns.lineplot(data = pop, x = 'Year', y = 'Value', hue = 'Country Name')
ylab = 'Population'
title = 'Population'
plt.ylabel(ylab)
plt.title(title)
```

## Steamlining the Code

While we can see the above code works well, there is a more streamlined way we can obtain the data for specific countries. This can be done by changing the indices of the rows on each table. With the below line, now the country code is the index.

```python
popData.index = popData['Country Code']
```

As you can see, the numerical index for each row has now been replaced with the Country Code.

```python
popData.head
```

Now using .loc with the Country Code, the data for a specific country or region can be obtained.

```python
popData.loc['ZWE']
```

To avoid the redundancy of the country code getting listed twice, the following code removes the display of the index.

```python
zwe = popData.loc['ZWE']
print(zwe.to_string(index = False))
```

```python
zwe.plot('Year', 'Value')
ylab = 'Population'
title = 'Zimbabwe Population'
plt.ylabel(ylab)
plt.title(title)
```

Now we can use this same idea to rewrite our code for plotting Spain, Zimbabwe, and Vietman on the same graph!

```python
pop = popData.loc[['ZWE', 'VNM', 'ESP']]
sns.lineplot(data = pop, x = 'Year', y = 'Value', hue = 'Country Name')
ylab = 'Population'
title = 'Population'
plt.ylabel(ylab)
plt.title(title)
```

## Filtering by Population Value

```python
highPop = popData.loc[(popData.Value > 1000000000) & (popData.Year == 1985)]
```

```python
highPop.plot('Country Name', 'Value', kind = 'bar')
ylab = 'Population'
title = 'Population greater than 1B'
plt.ylabel(ylab)
plt.title(title)
```

```python
zwePop = popData.loc[(popData['Country Name'] == 'Zimbabwe') & (popData.Value > 10000000)]
zwePop.plot('Year', 'Value', kind = 'bar')
ylab = 'Population'
title = 'Zimbabwe Population'
plt.ylabel(ylab)
plt.title(title)
```
