## Preliminary Steps
Import libraries needed for data analysis

```python
import matplotlib.pyplot as plt
import pandas as pd
```

Read in data from the csv file

```python
popData = pd.read_csv('../Downloads/population_csv.csv')
popData.head
```

## Creating a Line Plot
Create a DataFrame with Zimbabwe data

```python
zwe = popData.loc[popData['Country Name'] == 'Zimbabwe']
zwe
```

Plot Zimbabwe's population over time

```python
zwe.plot('Year', 'Value')
```

Add labels to make the plot more descriptive

```python
zwe.plot('Year', 'Value')
ylab = 'Population'
title = 'Zimbabwe Population'
plt.ylabel(ylab)
plt.title(title)
```
