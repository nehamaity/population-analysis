# Pharmaceutical Drug Spending by Country

In this tutorial we will using Python to conduct basic data analysis on a dataset showing pharmaceutical drug spending by country.

## Preliminary Steps

Import libraries needed for data analysis

```python
import matplotlib.pyplot as plt
import pandas as pd
```

Read in data from the csv file

```python
spendData = pd.read_csv('../Downloads/pharm_data_csv.csv', index_col = 0)
spendData.head
```

```python
spendData.describe()
```

## Australia Pharmaceutical Spending

Since we denoted the index as location, we can use the below code to create a data frame with all Australia data.

```python
aus = spendData.loc['AUS']
aus
```

We'll use a scatter plot to show the correlation between % of Health Spending with Spending in US GDP per capita

```python
aus.plot(kind = 'scatter', x = 'USD_CAP', y = 'PC_HEALTHXP')
xlab = 'Spending in US GDP per capita'
ylab = '% of Health spending'
title = 'Australia Pharmaceutical Spending'
plt.xlabel(xlab)
plt.ylabel(ylab)
plt.title(title)
```

We can include a little more information by making the Total Spending denoted by the size of the data points on the graph

```python
aus.plot(kind = 'scatter', x = 'USD_CAP', y = 'PC_HEALTHXP', s = aus['TOTAL_SPEND']/100)
xlab = 'Spending in US GDP per capita'
ylab = '% of Health spending'
title = 'Australia Pharmaceutical Spending'
plt.xlabel(xlab)
plt.ylabel(ylab)
plt.title(title)
```

Plotting the % of Heath spending against the year, we see a similar trend

```python
aus.plot(kind = 'scatter', x = 'TIME', y = 'PC_HEALTHXP', s = aus['TOTAL_SPEND']/100)
xlab = 'Year'
ylab = '% of Health spending'
title = 'Australia Pharmaceutical Spending'
plt.xlabel(xlab)
plt.ylabel(ylab)
plt.title(title)
```

Now we have a good idea of how the data trends in Australia. 

## 2005 Pharmaceutical Spending

Now we'll compare the spending of all countries in the year 2005 using the below code to obtain 2005 data.

```python
spend2005 = spendData.loc[(spendData.TIME == 2005)]
spend2005
```

```python
spend2005.plot.scatter(x = 'USD_CAP', y = 'PC_HEALTHXP', s = spend2005['TOTAL_SPEND']/100)
xlab = 'Spending in US GDP per capita'
ylab = '% of Health spending'
title = 'Pharmaceutical Spending in 2005'
plt.xlabel(xlab)
plt.ylabel(ylab)
plt.title(title)
```

The graph shows much more variation. To go a step further, we can add another piece of data from the table, the percentage of GDP (PC_GDP)

```python
spend2005.plot.scatter('USD_CAP', 'PC_HEALTHXP', s = spend2005['TOTAL_SPEND']/100, c = 'PC_GDP', 
                       colormap = 'viridis')
xlab = 'Spending in US GDP per capita'
ylab = '% of Health spending'
title = 'Pharmaceutical Spending in 2005'
plt.xlabel(xlab)
plt.ylabel(ylab)
plt.title(title)
```
